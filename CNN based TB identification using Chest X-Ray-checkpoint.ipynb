# Data Augmentation

```python
# Create Image Data Generator for Train Set
img_width = 150
img_height = 150

image_gen = ImageDataGenerator(
                                rescale = 1./255,
                                #shear_range = 0,
                                #zoom_range = 0,
                                horizontal_flip = True,          
                               )
# Create Image Data Generator for Test/Validation Set
test_data_gen = ImageDataGenerator(rescale = 1./255)
```

```python
# Creating the traing and testing set by using image data generator flow
train = image_gen.flow_from_directory(
      directory='C:/Users/Users/Somya/Image Analysis/dataset/train',
      target_size=(img_height, img_width),
      class_mode = 'sparse'
      )

test = image_gen.flow_from_directory(
      directory='C:/Users/Users/Somya/Image Analysis/dataset/test',
      target_size=(img_height, img_width),
      class_mode = 'sparse'
      )
```

# CNN Architecture

```python
#Setting the CNN model 
cnn = Sequential()

#convolution 
cnn.add(Conv2D(32, (3,3), input_shape = (img_width, img_height, 3), activation='relu', padding='same'))
cnn.add(MaxPool2D(2,2))

cnn.add(Conv2D(64, (3,3), activation='relu', padding='same'))
cnn.add(MaxPool2D(2,2))

cnn.add(Conv2D(128, (3,3), activation='relu', padding='same'))
cnn.add(MaxPool2D(2,2))

cnn.add(Conv2D(192, (3,3), activation='relu', padding='same'))
cnn.add(MaxPool2D(2,2))

#Dense 
cnn.add(Flatten())

cnn.add(Dense(128, activation='relu'))
cnn.add(Dropout(0.4))


cnn.add(Dense(228, activation='relu'))
cnn.add(Dropout(0.3))

cnn.add(Dense(270, activation='relu'))
cnn.add(Dropout(0.3))



cnn.add(Dense(1, activation='sigmoid'))
```

Model Compile

```python
cnn.compile(optimizer='Adam', loss='binary_crossentropy', metrics=['accuracy'])
```

Model Summary

```python
cnn.summary()
```

Visualize CNN model

```python
from tensorflow.keras.utils import plot_model
plot_model(cnn,show_shapes=True, show_layer_names=True, rankdir='TB', expand_nested=True)
```

Training


```python
#fitting the model on the training dataset
cnn.fit_generator(generator=train, 
                              steps_per_epoch=len(train),
                              epochs=20, 
                              validation_data= test, 
                              validation_steps = len(test))
```

# Analysis/Evaluation 

```python
#showing the graph of accuracy, validation accuracy, loss and validation loss for each epoch
pd.DataFrame(cnn.history.history).plot()
```

```python
#testing the model on test dataset
test_accu = cnn.evaluate(test)
print('The testing accuracy is :',test_accu[1]*100, '%')
```

```python
#find the prediction for the test dataset
preds = cnn.predict(test,verbose=1)
```

```python
predictions = preds.copy()
predictions[predictions <= 0.5] = 0
predictions[predictions > 0.5] = 1
```

# Results

```python
#showing the images with their prediction persentage
test.reset()
x=np.concatenate([test.next()[0] for i in range(test.__len__())])
y=np.concatenate([test.next()[1] for i in range(test.__len__())])
print(x.shape)
print(y.shape)
#this little code above extracts the images from test Data iterator without shuffling the sequence
# x contains image array and y has labels 
dic = {0:'NORMAL', 1:'TUBERCULOSIS'}
plt.figure(figsize=(20,20))
for i in range(0+10, 9+10):
  plt.subplot(3, 3, (i-10)+1)
  if preds[i, 0] >= 0.5: 
    out = ('{:.2%} probability of being Tuberculosis case'.format(preds[i][0]))
      
      
  else: 
    out = ('{:.2%} probability of being Normal case'.format(1-preds[i][0]))
  plt.title(out+"\n Actual case : "+ dic.get(y[i]))    
  plt.imshow(np.squeeze(x[i]))
  plt.axis('off')
plt.show()
```

```python
#finding the dice similarity coefficient
from sklearn.metrics import f1_score
from tensorflow.keras.metrics import binary_accuracy
dice_coefficient = f1_score(y_true=test.classes, y_pred=predictions, average='binary')
print("Dice similarity coefficient: {:.2f}".format(dice_coefficient))
```
